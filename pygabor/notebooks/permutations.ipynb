```python
import operator

from functools import reduce
from itertools import permutations, filterfalse
import numpy as np
```

```python
def fact(n):
    return reduce(operator.mul, list(range(n,0,-1)), 1)


def ilen(iterator):
    return reduce(lambda acc, _: acc + 1, iterator, 0)
```

```python
ilen(permutations(list(range(5,0,-1))))
```

```python
def restrict(n: int):
    sn = ilen(permutations(list(range(n,0,-1))))
    fn = ilen(filterfalse(lambda p: (1 in p[:4] or 2 in p[:4] or 3 in p[:4]), permutations(list(range(n,0,-1)))))
    return sn, fn

for n in range(7, 13):
    sn, fn = restrict(n)
    print(f'{n:>2}: {sn:>10} | {fn:<10} | {fact(n-4) * fact(n-3)}')
```

```python
FIRST = 4
N = 7

perms = permutations(list(range(N,0,-1)))
pos = np.zeros((N, FIRST))
for p in perms:
    print(f'{p}')
    for i in range(FIRST):
        pos[i][p[i] - 1] += 1
    for i in range(FIRST):
        print(f'\t{pos[i]}')

print(pos)
```

```python
def inv_restrict(n):
    sn = ilen(permutations(list(range(n,0,-1))))
    fn = ilen(filterfalse(lambda p: (1 not in p[:4] and 2 not in p[:4] and 3 not in p[:4]), permutations(list(range(n,0,-1)))))
    return sn, fn

for n in range(4, 11):
    sn, fn = inv_restrict(n)
    print(f'{n:>2}: {sn:>10} | {fn:<10} | {fact(n-4) * fact(n-3)}')
```
